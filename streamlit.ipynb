```python
import streamlit as st
```

```python
#Canvas
st.title("My first attempt on Streamlit")
#Layout(sidebar)
with st.sidebar:
    name = st.text_input("Enter your name") #Text Input
#logic and display
if name:
    st.write(f"Welcome to Streamlit, {name}!")
#canvas function
st.success("You have successfully created your first Streamlit app!")
```
